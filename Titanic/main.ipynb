# Data processing


```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('train.csv')
```

```python
names = data['Name']
# data = data.drop('Name', axis = 1)
```

```python
y = data['Survived']
```

```python
data['Sex'] = data['Sex'].map({'male': 1, 'female': 0})
```

```python
data = data.drop(['Survived', 'Name'], axis = 1)
```

```python
data.head()
```

```python
data = data.fillna(value = 0)
```

```python
for column in data.columns:
    print(column)
    print(data[column].value_counts(dropna = False))
    print()
```

```python
data = pd.get_dummies(data)
```

```python
data
```

```python
data.info()
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(data, y, test_size = 0.3, random_state = 11)
```

# Creating decision tree

```python
tree = DecisionTreeClassifier(random_state = 11)
params = {
    'max_depth': np.arange(2, 12),
    'min_samples_split': np.arange(2, 12),
    'max_features': np.arange(220, 241)
}
best_tree = GridSearchCV(tree, param_grid = params, n_jobs = -1, verbose = True)
```

```python
best_tree.fit(X_train, y_train)
```

```python
best_tree.best_params_
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(best_tree.predict(X_valid), y_valid)
```

```python
y_valid.value_counts()
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 7, n_jobs = -1)
```

```python
knn.fit(X_train, y_train)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logit = LogisticRegression(random_state = 11, C = 1, verbose = True)
```

```python
params = {
    'max_iter': np.arange(50, 80)
}
best_logit = GridSearchCV(logit, param_grid = params)
```

```python
best_logit.fit(X_train, y_train)
```

```python
best_logit.best_params_
```

```python
accuracy_score(best_logit.predict(X_valid), y_valid)
```
