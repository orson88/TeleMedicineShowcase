```python
import IPython
from scipy.io import wavfile
import scipy.signal
import numpy as np
import matplotlib.pyplot as plt
import librosa
%matplotlib inline
```

# Different kind of graphs showing noice frequencies

```python
import noisereduce as nr
# load data
rate, data = wavfile.read("mywav.wav")
data = np.array(data,dtype = float)
# select section of data that is noise
noisy_part = np.array(data,dtype = float)
# perform noise reduction
reduced_noise = nr.reduce_noise(audio_clip=data, noise_clip=noisy_part, prop_decrease=1.0, verbose=True)
```

# It's a graph of changed voice

```python
fig, ax = plt.subplots(figsize=(20,3))
ax.plot(reduced_noise)
```
